# Notebook 01: Data Exploration & EDA
Explore the synthetic crop dataset, understand distributions, and visualize relationships.

```python
import sys
sys.path.insert(0, '../backend')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate dataset
from ml.train import generate_synthetic_data
df = generate_synthetic_data(n_per_crop=200)
print(df.shape)
df.head()
```

```python
print('Dataset Info:')
print(df.info())
print('\nDescriptive Stats:')
df.describe()
```

```python
# Yield distribution per crop
fig, ax = plt.subplots(figsize=(14, 5))
crop_means = df.groupby('crop')['yield_kg_per_ha'].mean().sort_values(ascending=False)
crop_means.plot(kind='bar', ax=ax, color='steelblue', edgecolor='white')
ax.set_title('Average Yield per Crop (kg/ha)', fontsize=14)
ax.set_xlabel('')
ax.set_ylabel('yield (kg/ha)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap
num_cols = ['nitrogen','phosphorus','potassium','ph','organic_carbon',
            'temperature','rainfall','humidity','sunshine_hours','yield_kg_per_ha']
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(df[num_cols].corr(), annot=True, fmt='.2f', cmap='YlGn', ax=ax)
ax.set_title('Feature Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
# Yield vs. Rainfall scatter
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].scatter(df['rainfall'], df['yield_kg_per_ha'], alpha=0.2, s=4, color='steelblue')
axes[0].set_xlabel('Rainfall (mm)')
axes[0].set_ylabel('Yield (kg/ha)')
axes[0].set_title('Yield vs Rainfall')

axes[1].scatter(df['temperature'], df['yield_kg_per_ha'], alpha=0.2, s=4, color='coral')
axes[1].set_xlabel('Temperature (°C)')
axes[1].set_ylabel('Yield (kg/ha)')
axes[1].set_title('Yield vs Temperature')
plt.tight_layout()
plt.show()
```
