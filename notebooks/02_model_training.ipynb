```python
import sys
sys.path.insert(0, '../backend')

from ml.train import train
metrics = train()
print('\nTraining complete!')
print(f"Best model: {metrics['best_model']}")
print(f"Test MAE: {metrics['test_metrics']['MAE']} kg/ha")
print(f"Test R²: {metrics['test_metrics']['R2']}")
```

```python
import json
from pathlib import Path

metrics_path = Path('../backend/ml/models/metrics.json')
with open(metrics_path) as f:
    all_metrics = json.load(f)

import pandas as pd
results_df = pd.DataFrame(all_metrics['all_results']).T
results_df
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(8, 4))
models = list(all_metrics['all_results'].keys())
r2s = [all_metrics['all_results'][m]['R2'] for m in models]
colors = ['#16a34a' if m == all_metrics['best_model'] else '#94a3b8' for m in models]
bars = ax.bar(models, r2s, color=colors, edgecolor='white')
ax.set_ylim(0, 1)
ax.set_ylabel('R² Score')
ax.set_title('Model Comparison — R² Score (higher is better)')
for bar, val in zip(bars, r2s):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, f'{val:.4f}', ha='center', va='bottom', fontsize=11)
plt.tight_layout()
plt.show()
```
